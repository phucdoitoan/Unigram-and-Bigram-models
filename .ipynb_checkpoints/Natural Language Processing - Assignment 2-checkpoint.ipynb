```python
def train_test_uni_bigram():
    # create txt file without emtpy line
    with open('data/text_no-empty-line.txt', 'w') as outfile1:
        with open('data/text_in-one-line.txt', 'w') as outfile:
            with open('data/combined_text.txt', 'r') as infile:
                for i, line in enumerate(infile):
                    line = line.strip() # delete \n in the end
                    if len(line) == 0:
                        continue
                    else:
                        outfile1.write(line + '\n')
                        outfile.write(line + ' ')
                        words = line.split(' ')
                    #if i < 5:
                    #    print(line)
                    #    print(words)

    import string

    letters = string.ascii_letters + string.digits + ' -.' + "'" # put ',' for considering ',' in sentence
    print('Known char: \n', letters)

    unknown = set()
    with open('data/text_in-one-line.txt', 'r') as f:
        for i, line in enumerate(f):
            line = line.strip()
            for elem in line:
                if elem not in letters:
                    unknown.add(elem)
                    #unknown += elem
                    #print(i+1, elem)

    print('Unknown char: \n', unknown)

    def convert_unknown(char):
        if char == 'œ':
            return 'oe'
        elif char == 'ñ':
            return 'n'
        elif char == 'ç':
            return 'c'
        elif char == 'è':
            return 'e'
        elif char == 'ô':
            return 'o'
        elif char == 'é':
            return 'e'
        elif char == 'ö':
            return 'o'
        elif char == 'á':
            return 'a'
        elif char == 'æ':
            return 'ae'
        elif char == 'è':
            return 'e'
        elif char == 'ï':
            return 'i'
        elif char == 'ë':
            return 'e'
        elif char == '’':
            return "'"
        elif char == '—':
            return ' '
        elif char in '!?;': # possible symbols to end sentence -> to '.'
            return '.'
        else:
            return ''

    #[convert_unknown(c) for c in unknown]

    line_count = 0
    with open('data/processed-text.txt', 'w') as outf:
        with open('data/text_in-one-line.txt', 'r') as inf:
            line = inf.read()
            for c in unknown:
                line = line.replace(c, convert_unknown(c))
            sentences = line.strip().split('.')
            for i, sen in enumerate(sentences):
                if len(sen) <= 1:
                    continue
                # 'BT' is the start symbols
                outf.write('BTS ' + sen.strip() + '\n')  #put '.' for adding '.' in the end of sentence
                line_count += 1
    print('total lines: ', line_count)

    # divide for testing

    with open('data/processed-text.txt', 'r') as file:
        all_file = file.read()
        all_sentences = all_file.split('\n')
        print(len(all_sentences))

    import random

    random.shuffle(all_sentences)

    N = len(all_sentences)
    train_size = int(N*0.9)

    train_sentences = all_sentences[:train_size]
    test_sentences = all_sentences[train_size:]

    with open('data/processed-text_train.txt', 'w') as file:
        for sen in train_sentences:
            file.write(sen + '\n')

    with open('data/processed-text_test.txt', 'w') as file:
        for sen in test_sentences:
            file.write(sen + '\n')

    # convert char to number:
    def char2number(char):
        if char in '1qazQAZ':
            return '1'
        elif char in '2wsxWSX':
            return '2'
        elif char in '3edcEDC':
            return '3'
        elif char in '4rfvRFV':
            return '4'
        elif char in '5tgbTGB':
            return '5'
        elif char in '6yhnYHN':
            return '6'
        elif char in "7ujmUJM'": # include "'" -> '7'
            return '7'
        elif char in '8ikIK':
            return '8'
        elif char in '9olOL':
            return '9'
        elif char in '0pP-':  # include '-' -> '0'
            return '0'
        elif char in ' ':
            raise Exception("Please use anotate_word to anotate each word separeatedly or using anotate_sentence to anotate sentence! Containing '%s' in the sequence" %char)
        else:
            raise Exception("Unknown char: '%s'" %char)

    # convert words to number sequence according to predifined rules
    def anotate_word(word, only_pronun=False):
        if only_pronun:
            if word == 'BTS': # special start symbol
                return 'BTS'
            else:
                return ''.join([char2number(c) for c in word])
        else:
            if word == 'BTS': # special start symbol
                return 'BTS/BTS'
            else:
                return ''.join([char2number(c) for c in word]) + '/' + word

    def anotate_sentence(sentence, only_pronun=False):
        words = sentence.strip().split()
        #print(words)

        return ' '.join([anotate_word(word, only_pronun) for word in words])


    #print(anotate_word('Word'))
    #print(anotate_word('number'))
    #print(anotate_sentence("If I don't get the hydrophoby  It'll be because you was born to hang laughed the saloon-keeper"))

    # generating anotated-text from processed-text
    with open('data/anotated-text_train.txt', 'w') as outfile:
        with open('data/processed-text_train.txt', 'r') as infile:
            for sen in infile:
                anotated_sen = anotate_sentence(sen)
                outfile.write(anotated_sen + '\n')





    pair_dict = {}
    with open('data/anotated-text_train.txt', 'r') as file:
        for line in file:
            words = line.strip().split()
            for w in words:
                pair = tuple(w.split('/'))
                if pair in pair_dict.keys():
                    pair_dict[pair] += 1
                else:
                    pair_dict[pair] = 1

    print('pair dictionary: ', len(pair_dict))

    smooth_amount = 3 # best 3
    smoothed_pair_dict = {pair: freq+smooth_amount for pair, freq in pair_dict.items()}
    total_occ = sum(list(smoothed_pair_dict.values()))

    smoothed_pair_prob = {pair: freq/total_occ for pair, freq in smoothed_pair_dict.items()}

    #for w in sorted(smoothed_pair_prob, key=smoothed_pair_prob.get, reverse=True):
    #    print(w, '->', '%s'%smoothed_pair_prob[w])

    word_dict = {}
    with open('data/processed-text_train.txt', 'r') as file:
        for line in file:
            words = line.strip().split()
            for w in words:
                if w in word_dict.keys():
                    word_dict[w] += 1
                else:
                    word_dict[w] = 1

    print('word dictionary: ', len(word_dict))

    smooth_amount = 3 # best 3
    smoothed_word_dict = {word: freq+smooth_amount for word, freq in word_dict.items()}
    total_occ = sum(list(smoothed_word_dict.values()))

    smoothed_word_prob = {word: freq/total_occ for word, freq in smoothed_word_dict.items()}

    #for w in sorted(smoothed_word_prob, key=smoothed_word_prob.get, reverse=True):
    #    print(w, '->', '%s'%smoothed_word_prob[w])

    count = 0
    bound_fred = 2
    for _, freq in word_dict.items():
        if freq <= bound_fred:
            count += 1
    #print('num of word with no more than %s occurence: ' %bound_fred, count)
    #print('total num of words in dictinary: ', len(word_dict))

    pronun_word_dict = {}
    for (pro, w), prob in smoothed_pair_prob.items():
        try:
            if pro in pronun_word_dict.keys():
                if prob/smoothed_word_dict[w] > pronun_word_dict[pro][1]:
                    pronun_word_dict[pro] = (w, prob/smoothed_word_dict[w])
            else:
                pronun_word_dict[pro] = (w, prob/smoothed_word_dict[w])
        except Exception:
            print('la', pro, 'la', w, prob)


    def decode(pronun):
        try:
            word = pronun_word_dict[pronun][0]
        except KeyError:
            word = 'T_T'
        return word

    decode('5636')



    # testing with test file:
    total_word = 0
    total_correct = 0
    with open('data/processed-text_test.txt', 'r') as file:
        for i, sen in enumerate(file):
            sen = sen.strip()
            sen_code = anotate_sentence(sen, only_pronun=True)
            decoded_sen = ' '.join([decode(pro) for pro in sen_code.split()])

            words = sen.split()
            decoded_words = decoded_sen.split()
            total_word += len(words)
            total_correct += sum([w1 == w2 for w1, w2 in zip(words,decoded_words)])
            if i < 5: #300 and i > 295:
                print(sen)
                print(decoded_sen)
                print('\n')

    print('Test decoding acc: %s/%s = %.2f %%' %(total_correct, total_word, 100*total_correct/total_word))
    unigram_test_acc = 100*total_correct/total_word


    # testing with train file:
    total_word = 0
    total_correct = 0
    with open('data/processed-text_train.txt', 'r') as file:
        for i, sen in enumerate(file):
            sen = sen.strip()
            sen_code = anotate_sentence(sen, only_pronun=True)
            decoded_sen = ' '.join([decode(pro) for pro in sen_code.split()])

            words = sen.split()
            decoded_words = decoded_sen.split()
            total_word += len(words)
            total_correct += sum([w1 == w2 for w1, w2 in zip(words,decoded_words)])
            if i < 5: #300 and i > 295:
                print(sen)
                print(decoded_sen)
                print('\n')

    print('Train decoding acc: %s/%s = %.2f %%' %(total_correct, total_word, 100*total_correct/total_word))
    unigram_train_acc = 100*total_correct/total_word





    # make bi-gram

    word_pair_freq = {}

    with open('data/processed-text_train.txt', 'r') as file:
        for line in file:
            words = line.strip().split()
            words_num = len(words)
            for i in range(words_num - 1):
                pair = (words[i], words[i+1])
                if pair in word_pair_freq.keys():
                    word_pair_freq[pair] += 1
                else:
                    word_pair_freq[pair] = 1

    print('len word_pair_freq: ', len(word_pair_freq))



    word_bi_prob = {} # dict: (w2, w1) -> p(w2|w1)

    for (w1, w2) in word_pair_freq.keys():
        # p(w2|w1) = p(w1 w2) / p(w1)
        word_bi_prob[(w2, w1)] = word_pair_freq[(w1, w2)] / word_dict[w1]

    print('len word_bi_prob: ', len(word_bi_prob))

    # dict: y -> [w1, w2] such that .../w1 y/... in the corpus: 
    # y is the pronun of a word after w1, w2 ...
    pro2prev_w = {} 

    with open('data/anotated-text_train.txt', 'r') as file:
        for line in file:
            pairs = line.strip().split()
            pair_num = len(pairs)
            for i in range(0, pair_num-1):
                prev_w = pairs[i].split('/')[1]
                pronun = pairs[i+1].split('/')[0]
                if pronun in pro2prev_w.keys():
                    pro2prev_w[pronun].add(prev_w)
                else:
                    pro2prev_w[pronun] = set()
                    pro2prev_w[pronun].add(prev_w)

    print('len pro2prev_w: ', len(pro2prev_w))


    pronun_word_bi_dict = {} # dict: (y, w1) -> (w, p(w|y, w1))

    for (y, w), _ in smoothed_pair_prob.items():
        if w == 'BTS':
            pronun_word_bi_dict[y] = (w, 1)
        else:
            for w1 in pro2prev_w[y]:
                try:
                    p_w_w1 =  word_bi_prob[(w, w1)]
                except:
                    p_w_w1 = 0
                prob = (smoothed_pair_prob[(y, w)] / smoothed_word_prob[w]) * p_w_w1
                if (y, w1) in pronun_word_bi_dict.keys():
                    if pronun_word_bi_dict[(y,w1)][1] < prob: 
                        pronun_word_bi_dict[(y,w1)] = (w, prob)
                else:
                    pronun_word_bi_dict[(y,w1)] = (w, prob)

    print('len : ', len(pronun_word_bi_dict))

    def decode_bi(y, prev_w):
        if y == 'BTS':
            return 'BTS'
        else:
            try:
                return pronun_word_bi_dict[(y, prev_w)][0]
            except:
                if y in pronun_word_dict.keys():
                    return pronun_word_dict[y][0]
                else:
                    return 'T.T'


    # testing with test file:
    total_word = 0
    total_correct = 0
    with open('data/processed-text_test.txt', 'r') as file:
        for i, sen in enumerate(file):
            sen = sen.strip()
            sen_code = anotate_sentence(sen, only_pronun=True).split()
            decoded_words = []
            prev_w = ''
            for c in sen_code:
                prev_w = decode_bi(c, prev_w)
                decoded_words.append(prev_w)

            words = sen.split()
            correct = sum([w1 == w2 for w1, w2 in zip(words,decoded_words)])
            #if correct < len(words):
            #    print(sen)
            #    print(' '.join(decoded_words))

            total_word += len(words)
            total_correct += correct
            if i < 5: #300 and i > 297:
                print(sen)
                print(' '.join(decoded_words))
                print('\n')

    print('Test decoding acc: %s/%s = %.2f %%' %(total_correct, total_word, 100*total_correct/total_word))
    bigram_test_acc = 100*total_correct/total_word


    # testing with train file:
    total_word = 0
    total_correct = 0
    with open('data/processed-text_train.txt', 'r') as file:
        for i, sen in enumerate(file):
            sen = sen.strip()
            sen_code = anotate_sentence(sen, only_pronun=True).split()
            decoded_words = []
            prev_w = ''
            for c in sen_code:
                prev_w = decode_bi(c, prev_w)
                decoded_words.append(prev_w)

            words = sen.split()

            correct = sum([w1 == w2 for w1, w2 in zip(words,decoded_words)])
            #if correct < len(words):
            #    print(sen)
            #    print(' '.join(decoded_words))

            total_word += len(words)
            total_correct += correct
            if i < 5: #300 and i > 297:
                print(sen)
                print(' '.join(decoded_words))
                print('\n')

    print('Train decoding acc: %s/%s = %.2f %%' %(total_correct, total_word, 100*total_correct/total_word))
    bigram_train_acc = 100*total_correct/total_word

    return unigram_test_acc, unigram_train_acc, bigram_test_acc, bigram_train_acc
```

```python
import numpy as np

repeat_times = 50

unigram_test = []
unigram_train = []
bigram_test = []
bigram_train = []

for _ in range(repeat_times):
    uni_test, uni_train, bi_test, bi_train = train_test_uni_bigram()
    unigram_test.append(uni_test)
    unigram_train.append(uni_train)
    bigram_test.append(bi_test)
    bigram_train.append(bi_train)

unigram_test_np = np.asarray(unigram_test)
unigram_train_np = np.asarray(unigram_train)
bigram_test_np = np.asarray(bigram_test)
bigram_train_np = np.asarray(bigram_train)

```

```python
print('\n\nAverage unigram test: %.2f +- %.2f' %(np.mean(unigram_test_np), np.std(unigram_test_np)))
print('Average unigram train: %.2f +- %.2f' %(np.mean(unigram_train_np), np.std(unigram_train_np)))

print('\nAverage bigram test: %.2f +- %.2f' %(np.mean(bigram_test_np), np.std(bigram_test_np)))
print('Average bigram train: %.2f +- %.2f' %(np.mean(bigram_train_np), np.std(bigram_train_np)))
```
